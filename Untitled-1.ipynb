```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 10000

data = {
    'Tower_ID': np.random.randint(1000, 2000, n),
    'Region': np.random.choice(['Bangalore','Mysore','Hubli','Belgaum','Chikkmagaluru','Hassan','Mandy'], n),
    'Uptime_Percentage': np.random.uniform(90,100,n),
    'Downtime_Hours': np.random.uniform(0,20,n),
    'Signal_Strength': np.random.uniform(50,100,n),
    'Data_Traffic_GB': np.random.uniform(100,1000,n),
    'Maintenance_Cost': np.random.uniform(5000,50000,n),
    'Fault_Count': np.random.randint(0,10,n)
}

df = pd.DataFrame(data)

df.to_csv('telecom_tower.csv',index=False)

print(df.head())
```
