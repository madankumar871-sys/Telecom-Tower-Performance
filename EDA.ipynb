# Exploratory Data Analysis

Which region has maximum downtime?

```python
df.groupby('Region')['Downtime_Hours'].sum()
```

Which tower has most faults?

```python
df.groupby('Tower_ID')['Fault_Count'].sum()
```

Signal Strength Distribution

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Create a simple histagram of Signal_Strength
plt.figure(figsize=(10,6))
sns.histplot(df['Signal_Strength'],kde=True)
plt.title('Distibution of Signal_Strength in Telecom Tower Performance dataset')
plt.xlabel('Signal_Strength')
plt.ylabel('Tower_ID')
plt.axvline(df['Signal_Strength'].mean(),color='r',linestyle='--',label=f'mean:{df['Signal_Strength'].mean():.2f}')
plt.axvline(df['Signal_Strength'].median(),color='g',linestyle='-.',label=f'median:{df['Signal_Strength'].median():.2f}')
plt.legend()
plt.show()
```
